```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
#PONTOS DA SUSPENSÂO
#ua:upper arm
#la:lower arm
#t:tie-rod (Toe-Link)
#m:manga
#a:amortecedor

#pontos de chassi são: ua1,ua2,la1,la2 e a1

def converte(uaf,m1,uat,laf,m2,lat,t1,t2,a1,a2):
    lista=[uaf,m1,uat,laf,m2,lat,t1,t2,a1,a2]
    for i in lista:
        for a in range(len(i)):

            i[a-1]=i[a-1]/1000
            
    return lista


#pontos originais fx40
chassi_sup_frente=[-229.61,181.56,1104.84]
manga_sup=[-570.76,-3.3,935.89]
chassi_sup_tras=[-268.57,109.04,759.54]

chassi_inf_frente=[-244.01,81.77,1094.17]
manga_inf=[-582,-83.3,950]
chassi_inf_tras=[-277.85,11.09,777.81]

trod_chassi=[-235.55,111.22,1053.76]
trod_manga=[-596.17,-54.03,985.96]

amortecedor_chassi=[-236.81,367.51,916.65]
amortecedor_braco=[-426.51,27.26,935.18]
```

```python
def distancia(a,b):
    x=a[0]-b[0]
    y=a[1]-b[1]
    z=a[2]-b[2]
    resultado=math.sqrt(x**2+y**2)
    resultado=math.sqrt(resultado**2+z**2)
    return resultado
distancia(chassi_sup_frente,chassi_inf_tras)
print(distancia(manga_sup,chassi_sup_tras))
```
